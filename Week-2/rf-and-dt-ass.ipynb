# Random Forests and Decision Trees
- Classify the Spotify.csv dataset 
- Predict the Genre 
- The dataset has 23 Columns and The Output feature is **playlist_genre** so all others can be used as input features. Decide on what can be used and what cannot be used. 
- Do the Preprocessing below
- Do 70:30 Train test split 

```python
import pandas as pd 
df = pd.read_csv('/kaggle/input/spotify/spotify Edited.csv')
df.head()
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df.drop(columns = ['track_id', 'track_name', 'track_album_id','track_album_name', 'track_album_release_date'] , axis = 1)
```

```python
categoricalCols = df.select_dtypes(include=object).columns
numericalCols = df.select_dtypes(include=['int64', 'float64']).columns

cat = df[categoricalCols]
num = df[numericalCols]
le = LabelEncoder()

for i in cat.columns:
    cat.loc[:, i] = le.fit_transform(cat[i])
cat.head()
    
```

```python

scaler = StandardScaler()

scaled_num = scaler.fit_transform(num)
scaled_num = pd.DataFrame(scaled_num, columns = num.columns)
```

```python
scaled_num = pd.concat([scaled_num, cat], axis=1)
y = scaled_num.pop('playlist_genre').astype(int)
X = scaled_num
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Frame a Decision Tree Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini', max_depth = 20, random_state= 42)
```

## Frame a Random Forests Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=55, max_depth=55, random_state=4)
```

## Train the Models and Report the Accuracy

```python
dt.fit(X_train, y_train)
y_predDT = dt.predict(X_test)

rf.fit(X_train, y_train)
y_predRF = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"DT accuracy = {accuracy_score(y_true=y_test, y_pred=y_predDT)}")
print(f"RF accuracy = {accuracy_score(y_true=y_test, y_pred=y_predRF)}")
```
